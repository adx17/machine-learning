#**MEAN-SHIFT ALGORITHM**

##**IMPORTING MODULES**

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MeanShift,estimate_bandwidth
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

#**LOADING DATASET**

```python
data=load_iris()
X,Y=data.data,data.target
```

#**STANDARDISING THE DATA**

```python
x=StandardScaler().fit_transform(X) #Without scaling, features with larger numerical ranges dominate the clustering process.
```

#**ESTIMATING THE BANDWIDTH**

```python
bandwidth=estimate_bandwidth(x,quantile=0.3) #The function estimate_bandwidth() helps automatically determine a suitable bandwidth for Mean-Shift clustering.
                                              #low quantile results small bandwidth(more clusters)  [similar for high quantile]
```

##**FITTING THE MODEL**

```python
model=MeanShift(bandwidth=bandwidth)
model.fit(x)
labels=model.labels_
```

##**EVALUATING THE MODEL**

```python
ss=silhouette_score(x,labels)
print("Silhouette score:",ss)
```

##**DATA VISUALIZATION**

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap="rainbow",edgecolors='k')
plt.title("Mean-Shift Clustering on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
